```python
datos.to_csv('resultado_Time_Performance.csv')#se crea archivo csv
```

```python
#Carrier, FlightNum, Origin, Dest, total de flights por aerolínea y promedio de demoras por aeropuerto.
agrupados = datos[['DepDelay', 'DepDelayMinutes', 'DepartureDelayGroups', 'ArrDelay', 'ArrDelayMinutes'
                  , 'ArrivalDelayGroups', 'CarrierDelay', 'WeatherDelay', 'NASDelay'
                  , 'SecurityDelay', 'LateAircraftDelay', 'OriginAirportID']].groupby('OriginAirportID')
display(agrupados.mean().sort_values('DepDelay', ascending=False)) #agrupamos lo datos para ver el promedio de demora por aeropuerto
```

```python
#Obtener nombre de máquinas y aerolíneas respectivamente propensas a descomponerse y demorarse.
dems = datos[(datos['ArrDelay'] >0) | (datos['DepDelay'] >0) | (datos['CarrierDelay'] >0) 
           | (datos['WeatherDelay'] >0) | (datos['NASDelay'] >0) | (datos['SecurityDelay'] >0)
           | (datos['LateAircraftDelay'] >0)]
demoras = pd.Series.value_counts(dems['AirlineID']).to_dict()
```

```python
totales = pd.Series.value_counts(datos['AirlineID']).to_dict()
```

```python
for key in demoras:#for para ver el promedio de demoras por aerolinea
    print(str(key) + ": " + str(demoras[key]/totales[key]))
```

```python
titanic2 = pd.read_csv('../../titanic3.csv')
# Limpieza de datos 
titanic2["age"] = titanic2["age"].fillna(titanic2["age"].median())
titanic2["fare"] = titanic2["fare"].fillna(titanic2["fare"].median())

# Volvemos numericos los sexos.
titanic2.loc[titanic2["sex"] == "male", "sex"] = 0
titanic2.loc[titanic2["sex"] == "female", "sex"] = 1
# Volvemos numerico la columna de embarco
titanic2["embarked"] = titanic["embarked"].fillna("S")
titanic2.loc[titanic2["embarked"] == "S", "embarked"] = 0
titanic2.loc[titanic2["embarked"] == "C", "embarked"] = 1
titanic2.loc[titanic2["embarked"] == "Q", "embarked"] = 2
# Creamos la columna de familia
titanic2.loc[:,"family"] = titanic2.loc[:,"sibsp"] + titanic.loc[:,"parch"]
titanic2.loc[:,"farebin"] = (titanic2.loc[:,"fare"]/10.0).astype(int)
titanic2.loc[:,"agebin"] = (titanic2.loc[:,"age"]/10.0).astype(int)

# Definimos los predictores
predictors = ["pclass", "sex", "family", "agebin", "farebin"]
```
